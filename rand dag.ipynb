```python

import networkx as nx
import random
import numpy as np
from numpy.random import randint

np.random.seed(0)
def random_dag(nodes, edges):
    G = nx.DiGraph()
    for i in range(nodes):
        G.add_node(i)
    while edges > 0:
        a = randint(0,nodes-1)
        b=a
        while b==a:
            b = randint(0,nodes-1)
        G.add_edge(a,b)
        if nx.is_directed_acyclic_graph(G):
            edges -= 1
        else:
            # we closed a loop!
            G.remove_edge(a,b)
    return G
```

```python
from matplotlib import pyplot as plt
g = random_dag(1000 , 500 )
nx.draw(g, node_size = 50)
plt.show()


#todo clean up orphan nodes

#categorize node connectivity

#add graph chunk 

#connect 2 graphs


```
