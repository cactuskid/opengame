```python

import networkx as nx
import random
import numpy as np
from numpy.random import randint

np.random.seed(0)
def random_dag(G=None , nodes =10, edges=10):
    if not G:
        G = nx.DiGraph()
    for i in range(nx.number_of_nodes(G), nodes+nx.number_of_nodes(G)):
        G.add_node(i)
    while edges > 0:
        a = randint(0,nx.number_of_nodes(G)-1)
        b=a
        while b==a:
            b = randint(0,nx.number_of_nodes(G)-1)
        G.add_edge(a,b)
        if nx.is_directed_acyclic_graph(G):
            edges -= 1
        else:
            # we closed a loop!
            G.remove_edge(a,b)
    return G
```

```python
from matplotlib import pyplot as plt
#iteratively add to random DAG
g = random_dag(None, 100 , 100 )
for i in range (10):
    plt.clf()
    g = random_dag(g, 100 , 100 )
    nx.draw(g, node_size = 50)
    plt.show()

#todo clean up orphan nodes

#categorize node connectivity

#add graph chunk 

#connect 2 graphs?


```
